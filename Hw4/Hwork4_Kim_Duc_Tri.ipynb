Bài 1: Hãy xây dựng một danh sách liên kết đôi(DoubleLinked List –DLL) với các hàm theo yêu cầu sau:
1.1 Thêm vào một node có giá trịel ởcuối danh sách liên kết

1.2 Thêm vào một node có giá trịel ởđầu danh sách liên kết

1.3Thêm vào một node có giá trịel ởvịtrí pos trong danh sách

1.4Đếm sốphần tửtrong danh sách

1.5Xóa 1 phần tửởđầu danh sách

1.6Xóa 1 phần tửởcuối danh sách

1.7Xóa 1 phần tửởvịtrí pos trong danh sách

1.8Tìm một phần tửcó giá trịel và trảvềvịtrí của phần tửtrong danh sách, nếu không có trảvềgiá trị-1

1.9Tìm một phần tửởvịtrí pos trong danh sách và trảvềgiá trịcủa nó

1.10In ra danh sách 

1.11Sắp xếp danh sách theo giải thuật SelectionSort và in ra kết quả

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        """1.1 Thêm vào một node có giá trị el ở cuối danh sách liên kết"""
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        new_node.prev = last

    def push(self, data):
        """1.2 Thêm vào một node có giá trị el ở đầu danh sách liên kết"""
        new_node = Node(data)
        new_node.next = self.head
        if self.head is not None:
            self.head.prev = new_node
        self.head = new_node

    def insert_at(self, pos, data):
        """1.3 Thêm vào một node có giá trị el ở vị trí pos trong danh sách"""
        if pos == 0:
            self.push(data)
            return
        if self.head is None:
            return
        new_node = Node(data)
        count = 0
        temp = self.head
        while temp is not None and count < pos - 1:
            temp = temp.next
            count += 1
        if temp is None:
            return
        new_node.next = temp.next
        temp.next = new_node
        new_node.prev = temp
        if new_node.next is not None:
            new_node.next.prev = new_node

    def size(self):
        """1.4 Đếm số phần tử trong danh sách"""
        temp = self.head
        count = 0
        while temp:
            count += 1
            temp = temp.next
        return count

    def delete_first(self):
        """1.5 Xóa 1 phần tử ở đầu danh sách"""
        if self.head is None:
            return
        if self.head.next is None:
            self.head = None
            return
        self.head = self.head.next
        self.head.prev = None

    def delete_last(self):
        """1.6 Xóa 1 phần tử ở cuối danh sách"""
        if self.head is None:
            return
        if self.head.next is None:
            self.head = None
            return
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.prev.next = None

    def delete_at(self, pos):
        """1.7 Xóa 1 phần tử ở vị trí pos trong danh sách"""
        if self.head is None:
            return
        if pos == 0:
            self.delete_first()
            return
        count = 0
        temp = self.head
        while temp is not None and count < pos:
            temp = temp.next
            count += 1
        if temp is None:
            return
        if temp.next is not None:
            temp.next.prev = temp.prev
        if temp.prev is not None:
            temp.prev.next = temp.next

    def search(self, el):
        """1.8 Tìm một phần tử có giá trị el và trả về vị trí của phần tử trong danh sách, nếu không
        có trả về giá trị -1"""
        temp = self.head
        count = 0
        while temp is not None:
            if temp.data == el:
                return count
            count += 1
            temp = temp.next
        return -1

    def get_element_at(self, pos):
        """1.9 Tìm một phần tử ở vị trí pos trong danh sách và trả về giá trị của nó"""
        if self.head is None:
            return None
        temp = self.head
        count = 0
        while temp is not None:
            if count == pos:
                return temp.data
            count += 1
            temp = temp.next
        return None

    def print_list(self):
        """1.10 In ra danh sách"""
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()

    def selection_sort_and_print(self):
        """1.11 Sắp xếp danh sách theo giải thuật Selection Sort và in ra kết quả"""
        if self.head is None:
            return
        temp = self.head
        while temp:
            min = temp
            r = temp.next
            while r:
                if min.data > r.data:
                    min = r
                r = r.next
            temp.data, min.data = min.data, temp.data
            temp = temp.next
        self.print_list()



```

```python
if __name__ == "__main__":
    dll = DoublyLinkedList()
    dll.append(1)
    dll.append(3)
    dll.append(2)
    dll.append(5)
    dll.append(4)
    dll.push(6)
    dll.print_list()
    
    dll.push(0)
    dll.print_list()
    print("List after sorting")
    dll.selection_sort_and_print()
    
    print("list after inserting 10 at pos 3")
    dll.insert_at(3, 10)
    dll.print_list()
    dll.delete_first()
    dll.print_list()
    dll.delete_last()
    dll.print_list()
    dll.delete_at(2)
    dll.print_list()
    print(dll.search(10))
    print(dll.get_element_at(2))
    print(dll.size())
```
