```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # 1.1 Thêm vào cuối danh sách
    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        new_node.prev = last

    # 1.2 Thêm vào đầu danh sách
    def push(self, data):
        new_node = Node(data)
        new_node.next = self.head
        if self.head is not None:
            self.head.prev = new_node
        self.head = new_node

    # 1.3 Thêm vào vị trí pos
    def insert_at(self, pos, data):
        if pos == 0:
            self.push(data)
            return
        new_node = Node(data)
        count = 0
        temp = self.head
        while temp and count < pos - 1:
            temp = temp.next
            count += 1
        if temp is None:
            return
        new_node.next = temp.next
        new_node.prev = temp
        if temp.next:
            temp.next.prev = new_node
        temp.next = new_node

    # 1.4 Đếm số phần tử
    def size(self):
        count = 0
        temp = self.head
        while temp:
            count += 1
            temp = temp.next
        return count

    # 1.5 Xóa đầu danh sách
    def delete_first(self):
        if self.head is None:
            return
        if self.head.next is None:
            self.head = None
            return
        self.head = self.head.next
        self.head.prev = None

    # 1.6 Xóa cuối danh sách
    def delete_last(self):
        if self.head is None:
            return
        if self.head.next is None:
            self.head = None
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.prev.next = None

    # 1.7 Xóa ở vị trí pos
    def delete_at(self, pos):
        if pos == 0:
            self.delete_first()
            return
        count = 0
        temp = self.head
        while temp and count < pos:
            temp = temp.next
            count += 1
        if temp is None:
            return
        if temp.prev:
            temp.prev.next = temp.next
        if temp.next:
            temp.next.prev = temp.prev

    # 1.8 Tìm phần tử có giá trị el
    def find(self, el):
        count = 0
        temp = self.head
        while temp:
            if temp.data == el:
                return count
            count += 1
            temp = temp.next
        return -1

    # 1.9 Tìm phần tử ở vị trí pos
    def get_at(self, pos):
        count = 0
        temp = self.head
        while temp and count < pos:
            temp = temp.next
            count += 1
        if temp is None:
            return None
        return temp.data

    # 1.10 In danh sách
    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()

    # 1.11 Sắp xếp Selection Sort
    def selection_sort(self):
        current = self.head
        while current:
            min_node = current
            temp = current.next
            while temp:
                if temp.data < min_node.data:
                    min_node = temp
                temp = temp.next
            current.data, min_node.data = min_node.data, current.data
            current = current.next


# Ví dụ sử dụng
dll = DoublyLinkedList()

print("Danh sách rỗng:")
dll.print_list()  


dll.append(5)
dll.push(3)
dll.append(7)
dll.insert_at(2, 6)
print("Danh sách sau khi thêm phần tử:")
dll.print_list()  


print("Số lượng phần tử:", dll.size())  


dll.delete_at(1)
print("Danh sách sau khi xóa phần tử ở vị trí 1:")
dll.print_list() 


print("Vị trí của phần tử có giá trị 6:", dll.find(6))  
print("Vị trí của phần tử có giá trị 10:", dll.find(10))  


print("Giá trị của phần tử ở vị trí 2:", dll.get_at(2))  


dll.selection_sort()
print("Danh sách sau khi sắp xếp:")
dll.print_list()  
```

```python
class User:
    def __init__(self, nome, cognome):
        self.nome = nome
        self.cognome = cognome


class Subscriber(User):
    def __init__(self, nome, cognome, telefono, indirizzo, card):
        super().__init__(nome, cognome)
        self.telefono = telefono
        self.indirizzo = indirizzo
        self.card = card  


class Card:
    def __init__(self, credit, code):
        self.credit = credit
        self.code = code


class Movie:
    def __init__(self, name, type):
        self.name = name
        self.type = type  


class MovieRent:
    def __init__(self, movie, stato, data_prestito, prezzo):
        self.movie = movie
        self.stato = stato 
        self.data_prestito = data_prestito
        self.prezzo = prezzo  


class MovieBuy:
    def __init__(self, movie, stato, prezzo_vendita):
        self.movie = movie
        self.stato = stato  
        self.prezzo_vendita = prezzo_vendita 


class Order:
    def __init__(self, date, user, movies):
        self.date = date
        self.user = user  
        self.movies = movies  


class MovieShop:
    def __init__(self):
        self.movies = []  
        self.subscribers = []  #

    def listMovies(self):
        return self.movies

    def subscribe(self, user, card):
        """
        Add a new subscriber to the MovieShop
        """
        subscriber = Subscriber(user.nome, user.cognome, "N/A", "N/A", card)
        self.subscribers.append(subscriber)
        return subscriber

    def order(self, subscriber, movie, order_type):
        """
        Create an Order for the subscriber to either rent or buy a movie
        """
        from datetime import date

        if order_type == "rent":
            movie_rent = MovieRent(movie, "in", date.today(), 5.0)  #  rental price
            order = Order(date.today(), subscriber, [movie_rent])
            return order

        elif order_type == "buy":
            movie_buy = MovieBuy(movie, "presente", 20.0)  # purchase price
            order = Order(date.today(), subscriber, [movie_buy])
            return order

        else:
            raise ValueError("Invalid order type. Choose 'rent' or 'buy'.")


```

```python

if __name__ == "__main__":
    
    movie_shop = MovieShop()

    
    movie1 = Movie("The Shawshank Redemption", "DVD")
    movie2 = Movie("The Godfather", "VHS")
    movie_shop.movies.extend([movie1, movie2])

   
    user1 = User("John", "Doe")
    card1 = Card(credit=50, code="123456")
    subscriber1 = movie_shop.subscribe(user1, card1)

    print("Available movies:")
    for movie in movie_shop.listMovies():
        print(f"- {movie.name} ({movie.type})")

   
    rental_order = movie_shop.order(subscriber1, movie1, "rent")
    print(f"\nRental Order: {rental_order.user.nome} rented {rental_order.movies[0].movie.name}")

    
    purchase_order = movie_shop.order(subscriber1, movie2, "buy")
    print(f"\nPurchase Order: {purchase_order.user.nome} bought {purchase_order.movies[0].movie.name}")
```
